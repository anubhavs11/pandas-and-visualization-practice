### Outer Joins

```python
week1 = pd.read_csv("input/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("input/Restaurant - Week 2 Sales.csv")
foods = pd.read_csv("input/Restaurant - Foods.csv")
customer = pd.read_csv("input/Restaurant - Customers.csv")
```

```python
week1.merge(week2, how = "outer" , on = "Customer ID").tail()
# OUTER JOIN -> conatins all (both,left_only,right_only)
```

```python
week1.merge(week2, how = "outer" , on = "Customer ID",indicator = True).head()
```

```python
merged = week1.merge(week2, how = "outer" , on = "Customer ID",indicator = True)
merged["_merge"].head()
```

```python
full_outer = merged["_merge"].isin(["left_only","right_only"])
full_outer.head()
```

```python
merged[full_outer]
# FULL OUTER JOIN
```

```python
inner = merged[merged["_merge"] == "both"]
inner
#INNER JOIN
```

```python
left = merged[merged["_merge"].isin(["both","left_only"])]
left
#LEFT JOIN
```

```python
right = merged[merged["_merge"].isin(["both","right_only"])]
right
#RIGHT JOIN
```

### Merging by indexes with left_index and right_index parameters

```python
week1 = pd.read_csv("input/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("input/Restaurant - Week 2 Sales.csv")
foods = pd.read_csv("input/Restaurant - Foods.csv")
customer = pd.read_csv("input/Restaurant - Customers.csv")
```

```python
foods.head(3)
```

```python
customer.head(3)
```

```python
customer.set_index("ID",inplace = True)
```

```python
customer.head(3)
```

```python
week1.head(2)
```

```python
week1.merge(customer , how = "left" , left_on = "Customer ID" , right_index = True).head(3)
# use index of right column as right_on column when right_index = True
```

```python
customer.merge(week1 , right_on = "Customer ID", how= "left" ,left_index = True).head(3)
```

```python
week1.set_index("Customer ID").head(2)
```

```python
week1.merge(customer, how = "left", left_index = True , right_index = True).head(3)
```

```python
customer.merge(week1, how = "inner", right_index = True, left_index = True ).head(3)
```
