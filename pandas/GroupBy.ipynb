### The .agg() method

```python
fortune = pd.read_csv("input/fortune1000.csv",index_col=["Rank"])
sectors = fortune.groupby(["Sector","Industry"])
fortune.head(3)
```

```python
sectors.agg("mean").head(3)
# mean opertion on all numerical values column
```

```python
sectors.agg(["mean","sum"]).head(3)
# mean and sum opertion on all numerical values column
```

```python
sectors.agg(["mean","sum","size"]).head(3)
# mean , sum and size opertion on all numerical values column
```

```python
sectors.agg({"Revenue":"sum",
                "Profits":"mean",
                "Employees":"sum"}).head(3)
# operations will be applied on corresponding columns
```

```python
sectors.agg({"Revenue":["sum","mean","size"],
                "Profits":["sum","mean","size"],
                "Employees":["sum","mean","size"]}).head(3)
```

### Iterating through the groups

```python
fortune = pd.read_csv("input/fortune1000.csv",index_col=["Rank"])
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector , data in sectors:
    highest_revenue_company_in_a_sector = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_company_in_a_sector) 
    # does not have inplace option,so assignent operator is used
```

```python
df.head()
# dataframe having highest revenue companies from each sectors
```

```python
locations = fortune.groupby("Location")
locations
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for loc,data in locations :
    highest_revenue_company_in_a_city = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_company_in_a_city)
df.head()
# returns dataframe having highest revenue companies in a city
```
