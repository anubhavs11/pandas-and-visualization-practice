```python
import pandas as pd
import numpy as np
```

### Reshaping

```python
df = pd.DataFrame(np.arange(8).reshape(4,2) ,columns = pd.Index(["A","B"],name = "letters"),index = ["X","Y","Z","W"])
df
```

### Mapping

```python
df
```

```python
mapped = {1:"one",3:"three",5:"five",7:"seven"}
df["numbers"] = df["B"].map(mapped)
df
```

```python
df.index = df.index.map(str.lower)
```

```python
df
```

### Outliers

```python
df = pd.DataFrame(np.random.randn(1000,4))
df.head(3)
```

```python
df.tail(3)
```

```python
df.describe()
```

```python
df[(np.abs(df)>3).any(1)]
```

```python
df[(np.abs(df)>3).any(1)] = np.sign(df)*3
```

```python
df.describe()
# look at change in max and min
# outlier is set to -3 and +3
```
