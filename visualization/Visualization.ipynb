### Bar Charts

```python
stocks = data.DataReader("GOOG" , "morningstar" , start = "2004-01-01" , end = "2016-12-31")
stocks.head(3)
```

```python
def rankPrice(price):
    if(price <= 550):
        return "Poor"
    elif(price <650 ):
        return "Satisfactory"
    else:
        return "stelar"
```

```python
values = stocks["Close"].apply(rankPrice).value_counts()
```

```python
values
```

```python
type(values)
```

```python
plt.style.use("bmh")
values.plot(kind = "bar") # by default kind = line
```

```python
values.plot(kind = "barh") 
```

### Pie Charts

```python
values.plot(kind = "pie")
```

### Histograms

```python
apple = data.DataReader("AAPL" , "morningstar" , start = "2015-01-01", end = "2018-06-27")
apple.head(3)
```

```python
apple.reset_index(0,inplace=True)
```

```python
apple.head(3)
```

```python
def custom_values(price):
    return int(price/100)*100
```

```python
stocks.Close = stocks.Close.apply(custom_values)
```

```python
values.plot(kind = "hist" )
```
