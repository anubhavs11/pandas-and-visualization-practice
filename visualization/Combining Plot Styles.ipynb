```python
import numpy as np
import pandas as pd

from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.distplot(dataset , bins = 25)
```

By default sns.distplot() combines kernel density estimation plot

```python
sns.distplot(dataset,bins = 25, hist = False , rug= True) # kde is by default True
```

```python
sns.distplot(dataset , bins = 25,
             kde_kws ={"color":'indianred','label':'KDE PLOT' },
            hist_kws = {'color':'blue' , 'label':'HIST'})
```

```python
ser1 = pd.Series(dataset , name= "Series")
```

```python
sns.distplot(ser1 , bins = 25)
```
